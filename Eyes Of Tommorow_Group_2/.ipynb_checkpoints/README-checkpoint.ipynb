## Cartoonize Image Using OpenCV

To cartoonize an image using OpenCV, you can follow these steps:

1. **Convert to Grayscale:** Convert the loaded image to grayscale. This step simplifies the image and helps in reducing noise.

2. **Apply Median Blur:** Apply median blur to the grayscale image. Median blur helps in smoothing the image while preserving edges.

3. **Detect Edges:** Use adaptive thresholding to detect edges in the image. Adaptive thresholding is a technique used to binarize images based on local pixel intensities, which helps in highlighting edges.

4. **Cartoonize the Image:**
   - **Bilateral Filtering:** Apply bilateral filtering to the original color image. Bilateral filtering smooths the image while preserving edges and texture.
   
   - **Bitwise AND Operation:** Perform a bitwise AND operation between the color-filtered image and the edge-detected image. This operation retains the color information from the original image while emphasizing the detected edges.

5. **Save and Display the Cartoonized Image:**
   - **Save Image:** Save the cartoonized image to a file using `cv2.imwrite()`.
   
   - **Display Image:** Display the cartoonized image in the Jupyter Notebook using `display(Image())`.

By following these steps, you can create a cartoonized version of any image using OpenCV.
